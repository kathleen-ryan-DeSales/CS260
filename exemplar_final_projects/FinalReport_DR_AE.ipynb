### 2.1 Variables Explained

The variables in the **geocoded** table are explained below.

* **observed:** In this column, the stories of the sightings go into detail of what happened
* **location_details:** In this column, the witnesses give detail of where the sighting was
* **county:** This column is a list of the counties the sightings were in
* **state:** This column is a list of all 50 states within the USA
* **season:** In this column, it tells us what season the sightings was in
* **title:** This column in the report numbers along with a description of the sightings; For example: *Report 31940: Fisherman and Law Enforcement Officer Report Three Class A Sightings in Same Area Within A Year near the town of Somers*
* **latitude:** This column describes the latitude of where the witness was at the time of the sightings
* **longitude:** This column describes the longitude of where the witness was at the time of the sightings
* **date:** This column give the date of the sightings
* **number:** This column is the ID number associated with the report
* **classification:** This column is used to identify the class the sightings were; The classes range from A - C, A being the clearest sightings with the most evidence and C being harder to tell if they are telling the truth or not; This coulmn is also very good for joining the different tables together. More info can be found here: https://www.bfro.net/gdb/classify.asp
* **geohash:** This column tells the code for the public domain geocode system
* **temperature_high:** This gives us the temperature high for the date of the sightings
* **temperature_mid:** This gives us the temperature mid for the date of the sightings
* **temperature_low:** This gives us the temperature low for the date of the sightings
* **dew_point:** This was what the dew point was at on the date of the sightings
* **humidity:** This column gives us the humidity number was at on the date of the sightings
* **cloud_cover:** This column explains how couldy the sky was on the date of the sightings; 0 having no clouds and 10 being the cloudiest
* **moon_phase:** This column tells us the phase of the moon on the date of the sightings
* **precip_intensity:** This column tells us what the intensity of the precipitation was on the date of the sightings
* **precip_probability:** This column tells us what the probability of the precipitation was on the date of the sightings
* **precip_type:** This column tells us what the type of the rainfall was on the date of the sightings; Whether it be rain, snow, or nothing at all
* **pressure:** This explains the pressure of the precipitation on the date of the sightings
* **summary:** This column explains what the weather was like on the date of the sightings; For example: *Mostly cloudy throughout the day*
* **uv_index:** This column gives the UV Index; A UV Index predicts the ultraviolet radiation levels on a 1-11+ scale and provides a daily forecast of the expected intensity of UV
* **visibilty:** This column tells us the visibility of the sightings
* **wind_bearing:** This column describes the wind direction
* **wind_speed:** In this column, it describes the speed the wind was blowing on the date of the sightings

All of the data in the **geocoded** table from observed to classification are qualitative except latitude, longitude and date. All of the data after classification is quantitative except summary which is qualitative. 

The variables in the **locations** table are explained below.
* **number:** This column is the ID number associated with the report
* **title:** This column in the report numbers along with a description of the sighting; For example: *Report 451: Young man has sighting near home*
* **classification:** This column is used to identify the class the sighting was; The classes range from A - C, A being the clearest sighting with the most evidence and C being harder to tell if they are telling the truth or not; This coulmn is also very good for joining the different tables together
* **timestamp:** This column tells us the day, year, month, and time of day of the sightings
* **latitude:** This column describes the latitude of where the witness was at the time of the sightings
* **longitude:** This column describes the longitude of where the witness was at the time of the sightings

All of the data in the **locations** table is quantitative with the exception of title and classification which are qualitative. 

The variables in the **us_pop** table are explained below.
* **date:** This column is the last date of every year 
* **population:**  In this column, the population of the US at that time
* **annual_%_change:** This column is the percentage change in population from year to year
