## 1. Introduction

The Waterfront Partnership of Baltimore launched the Healthy Harbor Initiative to help clean up the cityâ€™s waterways in an attempt to make them more enjoyable for both residents and visitors. The program focuses on restoration, education, and community involvement to improve water quality and increase public awareness of waterway pollution.

As part of that effort, Mr. Trash Wheel was installed in 2014 at the mouth of the Jones Falls River. Over the years, the project expanded with the addition of three more trash wheels: Professor Trash Wheel in 2016, Captain Trash Wheel in 2018, and Gwynnda the Good Wheel of the West in 2021. These devices are both functional and engaging, as they draw attention through their quirky names and designs.

This report looks at data collected from all four trash wheels across several years. The dataset includes information about the contents of each dumpster, the amount and type of garbage collected, and when the collection happened. We also used local traffic volume data to see if there is any connection between how busy the area is and how much trash ends up in the water.

Our analysis shows that the total amount of garbage collected has gone up over time, especially after each new trash wheel was installed. There is a noticeable spike in 2021 after Gwynnda was deployed. That does not necessarily mean pollution got worse. It could also mean the project is doing a better job at catching trash before it spreads.

A strong link between traffic volume and trash collected was found. Busier roads might be leading to more litter, which ends up in the harbor. Even though cigarette butts were the most common item found, we focused on plastic bottles in our analysis because they are widely known to cause long-term environmental problems. Bottle collection rates varied across the different wheels, giving us a look at how trash types differ by location. We also saw a connection between the number of wrappers and plastic bags, which could mean that some people are tossing full bags of trash into the water rather than individual pieces.

Mr. Trash Wheel, Professor Trash Wheel, Captain Trash Wheel, and Gwynnda the Good Wheel of the West are all located within the Baltimore Harbor area in Baltimore, Maryland USA. Mister Trash Wheel is located at the Jones Falls River mouth. Professor Trash Wheel is located at Harris Creek. Captain Trash Wheel is located at Masonville Cove Environmental Education Center. Gwynnda Trash Wheel is located at Ridgely's Cove at the mouth of the Gwynns Falls.

Since the inception of Mister Trash Wheel in May 2014, each dumpster load is either hand-counted or calculated with random samples per paddle. If crew members are present, they manually count the number of each type of trash listed on a single conveyor paddle. This process is repeated several times during the filling of a dumpster. An average is then calculated for each item per paddle. The average is then multiplied by the paddle rate and the time it takes to fill the dumpster. The dumpsters are counted at least once per year by volunteers and are otherwise calculated by random batch samples. The process is explained next.

Sometimes, volunteers hand-count the entire dumpster, though this cannot always be done due to the lengthy workload involved. If no crew is present, nor are there any volunteers, random bushel-sized samples are taken of the collected material. These samples are collected and multiplied by 325, as there are 325 bushels in a typical dumpster load. Professor Trash Wheel was introduced in December 2016, Captain Trash Wheel in June 2018, and Gwynnda Trash Wheel in June 2021. Their data in the CSV file reflects this accordingly.

## 4. Conclusion

Upon evaluating this dataset, there has not been a significant decrease in the amount of garbage collected. In fact, the overall amount of waste seems to be increasing. However, this does not indicate that the Healthy Harbor Initiative has been unsuccessful. The deployment of additional trash wheels likely contributes to the increased waste collection. The prevalence of plastic bottles and the strong correlation between wrappers and plastic bags may indicate that people are using bags to discard multiple garbage items. More widespread use of biodegradable packaging could help reduce the long-term environmental impact of this type of litter.

If a more comprehensive dataset were available, one with more collections per wheel, potentially in the future, it would allow for a clearer picture of long-term trends. Since the effects of the trash wheels may take years to appear, additional data points would strengthen our conclusions.
